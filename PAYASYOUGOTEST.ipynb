```python
Count_of_Transac=[]
t=range(28007,37343)
for i in or_test['TransactionDates'][t]:
    Count_of_Transac.append(len(i))
or_test['Count_of_Transac']=Count_of_Transac
```

```python
Max_Payment_hist=[]
t=range(28007,37343)
for i in or_test['PaymentsHistory'][t]:
    Max_Payment_hist.append(max(i))
or_test['Max_Payment_hist']=Max_Payment_hist
```

```python
Min_Payment_hist=[]
t=range(28007,37343)
for i in or_test['PaymentsHistory'][t]:
    Min_Payment_hist.append(min(i))
or_test['Min_Payment_hist']=Min_Payment_hist
```

```python
Median_Payment_hist=[]
t=range(28007,37343)
for i in or_test['PaymentsHistory'][t]:
    Median_Payment_hist.append(np.median(i))
or_test['Median_Payment_hist']=Median_Payment_hist
```

```python
Sum_Payment_hist=[]
t=range(28007,37343)
for i in or_test['PaymentsHistory'][t]:
    Sum_Payment_hist.append(sum(i))
or_test['Sum_of_Payment_hist']=Sum_Payment_hist
```

```python
or_test['Left_over_in_contract']=or_test['TotalContractValue']-or_test['Sum_of_Payment_hist']
```
